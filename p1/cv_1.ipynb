```python
import cv2
from matplotlib import pyplot as plt
```

Reading Image

```python
img = cv2.imread('opm.jpeg')
name = 'One Punch Man Image'
plt.title(name)
plt.imshow(img)
```

Flipping Horizontally

```python
h_flip = cv2.flip(img, 1)
name = 'Horizontal Flip'
plt.title(name)
plt.imshow(h_flip)
```

Flipping Vertically

```python
v_flip = cv2.flip(img, 0)
name = 'Vertical Flip'
plt.title(name)
plt.imshow(v_flip)
```

Flipping Vertically and Horizontally

```python
hv_flip = cv2.flip(img, -1)
name = 'Horizontal and Vertical Flip'
plt.title(name)
plt.imshow(hv_flip)
```
