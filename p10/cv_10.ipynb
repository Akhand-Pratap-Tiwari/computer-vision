# Corner Detection Using Harris

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('stones.png')
img_gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
img_gray = np.float32(img_gray)
plt.imshow(img)
```

```python
corner_img = cv2.cornerHarris(img_gray, 5, 5, 0.07)
corner_img = cv2.dilate(corner_img, None)
img[corner_img > 0.01 * corner_img.max()] = [255, 0, 0]
plt.imshow(img)
```
