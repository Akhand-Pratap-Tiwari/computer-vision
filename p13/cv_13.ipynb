# Image segmentation using K-Medoids

```python
import numpy as np
import seaborn as sns

from matplotlib import pyplot as plt
from sklearn_extra.cluster import KMedoids
```

```python
# DUmmy Data
data = np.array([[7, 8], [9, 10], [11, 5], [4, 9], [7, 5], [2, 3], [4, 5]])
```

```python
# Visualizaing the data set
plt.figure(figsize=(10, 8))
sns.scatterplot(x = data[:, 0], y = data[:, 1], markers='o', s=100)
plt.title('Dummy image data points')
plt.show()
```

```python
centers = [[4, 5], [9, 10]]
```

```python
Kmeds = KMedoids(n_clusters=2)
Kmeds.fit(data)

# Assigning the labels
labels = Kmeds.labels_
```

```python
# Vizualizing the clusters that are formed
unique_labels = set(labels)

# Color codes
colors_plot = [plt.cm.Spectral(each) for each in np.linspace(0, 1, len(unique_labels))]

xy = data

plt.figure(figsize=(10, 8))
sns.scatterplot(x = xy[:, 0], y=xy[:, 1], marker='+', hue=labels, s=100)

sns.scatterplot(x = Kmeds.cluster_centers_[:, 0], y = Kmeds.cluster_centers_[:, 1], markers='o', color = 'red', label='Centroids', s=100)

plt.title('KMedoids clustering on Dummy Data', fontsize=14)
plt.legend()
plt.show()
```
