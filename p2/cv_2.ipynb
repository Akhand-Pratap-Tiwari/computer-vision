# How to take bitwise_not pixel wise

```python
import cv2
from matplotlib import pyplot as plt
from PIL import Image
from numpy import array
```

```python
img = cv2.imread('opm.jpeg')
plt.imshow(img)
```

```python
img_bn = cv2.bitwise_not(img)
plt.imshow(img_bn)
cv2.imwrite('output.jpeg', img_bn)
```

```python
im1 = Image.open('opm.jpeg')
ar1 = array(im1)
display(ar1)

im2 = Image.open('output.jpeg')
ar2 = array(im2)
display(ar2)
```
