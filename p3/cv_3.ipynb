# How to read, save, display and perform basic operations

```python
from matplotlib import pyplot as plt
import cv2
```

## Reading Image 

```python
img = cv2.imread('opm.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.title('Original')
plt.imshow(img_rgb)
```

## Turning to Gray and Saving Image

```python
img_gray = cv2.imread('opm.jpeg', cv2.COLOR_BGR2GRAY)

cv2.imwrite('opm_gray.jpeg', img_gray)

plt.title('Gray Image')
plt.imshow(img_gray)
```

## Reading Saved Image

```python
img = cv2.imread('opm_gray.jpeg')

plt.title('Saved Image')
plt.imshow(img)
```
