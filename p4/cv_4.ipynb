# Image processing with OpenCV and Python

```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
from  numpy import array
```

```python
src = 'opm.jpeg'
input_img = cv2.imread(src)

cv2.imshow('Original Image', input_img)

cv2.waitKey(0)
cv2.destroyAllWindows()

if input_img is None:
    print('Could Not open the image', input_img)
    exit(0)

plt.title('Original Image')
plt.imshow(cv2.cvtColor(input_img, cv2.COLOR_BGR2RGB))
```

## Converting to gray scale

```python
img_gray = cv2.cvtColor(input_img, cv2.COLOR_BGR2GRAY)

cv2.imshow('Grayscale Image', img_gray)

cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Gray-Scale Image')
plt.imshow(cv2.cvtColor(img_gray, cv2.COLOR_BGR2RGB))
```

## Separating and Combining BGR Channels

```python
b, g, r = cv2.split(input_img)

cv2.imshow('Blue-Gray Scale', b)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Green-Gray Scale', g)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Red-Gray Scale', r)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Blue-Gray Scale Image')
plt.imshow(cv2.cvtColor(b, cv2.COLOR_BGR2RGB))
plt.figure()

plt.title('Green-Gray Scale Image')
plt.imshow(cv2.cvtColor(g, cv2.COLOR_BGR2RGB))
plt.figure()

plt.title('Red-Gray Scale Image')
plt.imshow(cv2.cvtColor(r, cv2.COLOR_BGR2RGB))
```

```python
b_channel = np.zeros(input_img.shape, input_img.dtype)
g_channel = np.zeros(input_img.shape, input_img.dtype)
r_channel = np.zeros(input_img.shape, input_img.dtype)

display(b_channel.shape, b.shape)

cv2.mixChannels([b, g, r], [b_channel], [0, 0])
cv2.mixChannels([b, g, r], [g_channel], [1, 1])
cv2.mixChannels([b, g, r], [r_channel], [2, 2]) 

cv2.imshow('Blue Channel', b_channel)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Green Channel', g_channel)
cv2.waitKey(0)
cv2.destroyAllWindows()

cv2.imshow('Red Channel', r_channel)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Blue Channel')
plt.imshow(cv2.cvtColor(b_channel, cv2.COLOR_BGR2RGB))
plt.figure()

plt.title('Green Channel')
plt.imshow(cv2.cvtColor(g_channel, cv2.COLOR_BGR2RGB))
plt.figure()

plt.title('Red Channel')
plt.imshow(cv2.cvtColor(r_channel, cv2.COLOR_BGR2RGB))
```

## Blurring an Image

```python
blur_img = cv2.medianBlur(input_img, 13)
cv2.imshow("Blur Image", blur_img)
cv2.waitKey(0)
cv2.destroyAllWindows()

plt.title('Blur Image')
plt.imshow(cv2.cvtColor(blur_img, cv2.COLOR_BGR2RGB))
```
