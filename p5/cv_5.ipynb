```python
import random
from matplotlib import pyplot as plt
import cv2
```

# Sample Image


# Function for adding noise

### Function add_noise() will add random black and white pixels in the image and will create salt and pepper effect on the image

```python
def add_noise(img):
    row, col = img.shape

    # For random sprinkle of white pixels
    no_of_pixels = random.randint(200, 1000)
    for i in range(no_of_pixels):
        y_cord = random.randint(0, row - 1)
        x_cord = random.randint(0, col - 1)
        img[y_cord][x_cord] = 255

    # For random sprinkle of black pixels
    no_of_pixels = random.randint(200, 1000)
    for i in range(no_of_pixels):
        y_cord = random.randint(0, row - 1)
        x_cord = random.randint(0, col - 1)
        img[y_cord][x_cord] = 0
    
    return img
    
```

## Reading Image

```python
img = cv2.imread('opm.jpeg')
plt.imshow(img)
plt.show()
```

## Reading in RGB rather than BGR

```python
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img_rgb)
plt.show()
```

## Salt Pepper effect addition

```python
img_gray = cv2.imread('opm.jpeg', cv2.IMREAD_GRAYSCALE)
cv2.imwrite('opm_with_noise.jpeg', add_noise(img_gray))
img_noise = cv2.imread('opm_with_noise.jpeg')
plt.imshow(img_noise)
plt.show()
```

## Removing the effect using median blur

```python
img_blur = cv2.medianBlur(img_noise, 3)
plt.imshow(img_blur)
plt.show()
```
