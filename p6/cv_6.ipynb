```python
import numpy as np
from matplotlib import pyplot as plt
import cv2
```

```python
def plotImageComparison(original_image, transformed_image, transformation_name):
    fig = plt.figure(figsize=(15, 5))

    ax = fig.add_subplot(121)
    plt.title('Original', fontsize=18)
    plt.imshow(original_image, cmap='gray')
    
    ax = fig.add_subplot(122)
    plt.imshow(transformed_image, cmap='gray')
    ax.set_title(transformation_name, fontsize=18)
```

# Linear Transformations

```python
img = cv2.imread('opm.jpeg', 0)
img_inv = 255 - img

plotImageComparison(img, img_inv, 'Inverse Transform')
```

# Log Transformation

```python
log = 0.6 * (np.log(1 + np.float32(img)))
plotImageComparison(img, log, 'Log Transform')
```

# Power Law (Gamma) Transformation

```python
img_1 = img/255.0
img_power_law = cv2.pow(img_1, 1.8)
plotImageComparison(img, img_power_law, 'Power Law Transform')
```
