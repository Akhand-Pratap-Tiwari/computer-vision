```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
def plotImageComparison(img1, name1, img2, name2):
    fig = plt.figure(figsize=(15, 5))

    ax = fig.add_subplot(121)
    plt.title(name1, fontsize=18)
    plt.imshow(img1, cmap='gray')
    
    ax = fig.add_subplot(122)
    plt.imshow(img2, cmap='gray')
    ax.set_title(name2, fontsize=18)
```

```python
img = cv2.imread('j.png')
img2 = 255-img
kernel = np.ones((7, 7), np.uint8)
plotImageComparison(img, 'Black J on white bg', img2, 'White J on black bg')
```

# Erosion Operation

```python
erosion = cv2.erode(img, kernel)
erosion2 = cv2.erode(img2, kernel)
plotImageComparison(erosion, 'Erosion Pumps J', erosion2, 'Erosion Shrinks J')
```

# Dilation Operation

```python
dilation = cv2.dilate(img, kernel)
dilation2 = cv2.dilate(img2, kernel)
plotImageComparison(dilation, 'Dilation Shrinks J', dilation2, 'Dilation Swells J')
```

### Note: If the J is black and background is white then on erosion J will swell while on dilation J will shrink. Reverse happens when J is in white and background is black. Hence, we must keep the object in white [object to be eroded or dilated] and bg in black for convenience.
