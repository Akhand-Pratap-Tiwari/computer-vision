# Edge Detection

```python
import cv2
from matplotlib import pyplot as plt
```

## Image Reading and Blurring (to reduce noise and unwanted stuff)

```python
img = cv2.imread('opm.jpeg')
imgRGB = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
imgGRAY = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
imgBLUR = cv2.GaussianBlur(imgGRAY, (3, 3), 0)
plt.title("Original Image")
plt.imshow(imgRGB)
```

```python
plt.title("Blurred Image")
plt.imshow(cv2.cvtColor(imgBLUR, cv2.COLOR_GRAY2RGB))
```

## Using Canny from openCV

```python
edges = cv2.Canny(image=imgBLUR, threshold1=100, threshold2=200)
edges = cv2.cvtColor(edges, cv2.COLOR_BGR2RGB)
plt.title("Edges")
plt.imshow(edges)
```
